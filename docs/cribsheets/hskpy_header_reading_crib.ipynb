# Get header values from fits data

```python
from hskpy.general.data import get_fname, HskData, get_header_value
## Read data
fname = get_fname('saturn', '20140610')
hskdat = HskData(fname)
```

## you can get any header value

```python
## you can get any header value using get_header_value function, but you do have to remember the name of header.
calflg = get_header_value(hskdat.hdul, 'CALFLG')
print(calflg)

## specify extensions
ext = [2,3,4,5,6]
calflg = get_header_value(hskdat.hdul, 'CALFLG', ext=ext)
print(calflg)
```

```python
## The same can be done using get_calflg function
from hskpy.general.data import get_calflg
calflg = get_calflg(hskdat.hdul)
print('string version', calflg)

## if you want calflg in interger format, make string keyword False
calflg_i = get_calflg(hskdat.hdul, string=False)
print('integer version', calflg_i)

## Can do the same using get_calflg method of HskData class
calflg_o = hskdat.get_calflg()
print('get from class',calflg_o)
```
